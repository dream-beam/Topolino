```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
dx_m = 1e-3 #grid size, m:meter
Dx_m = 1 #screen size
x_m = np.linspace(-Dx_m/2, Dx_m/2, int(Dx_m/dx_m)+1)
```

```python
print(x_m)
```

```python
y = x_m**2
plt.plot(x_m, y)
plt.xlabel("Position in meters")
plt.ylabel("Dummy variable")
plt.title("Quadratic function")
plt.show()
```

```python
f_pm = 5 #pm:permeter, screen is 1 meter
s = np.sin((2*np.pi*x_m)*f_pm)
plt.plot(x_m,s)
plt.xlabel("Position in meters")
plt.ylabel("Dummy variable 2")
plt.title("Sinusoidal function of frequency 10 Hz")
plt.show()
```

```python
x_mean_m = 0.1 #centered on...
x_std_m = 0.05 #widthofgaussian
g = np.exp(-(x_m - x_mean_m)**2/(2*(x_std_m**2)))
plt.plot(x_m, g)
plt.xlabel("Position in meters")
plt.ylabel("Intensity in au") #arbitrary units
plt.title("Gaussian function")
plt.show()
#f'n, not distribution
```

```python
x_mean_estimate_m = np.sum(x_m*g)/np.sum(g)#moment: sum of values * their position / sum of all values
print(x_mean_estimate_m)
```

```python
x_std_estimate_m = np.sqrt(np.sum((x_m - x_mean_estimate_m)**2*g)/np.sum(g))
print(x_std_estimate_m)
```

```python
n = np.random.randn(1000)
plt.plot(n)
```

```python
plt.hist(n)
plt.xlabel("Bin")
plt.ylabel("Count")
plt.title("Histogram of a Gaussian Distribution")
plt.show()
#actual gaussian, not f'n
```

```python
np.mean(n) #should be close to 0
np.std(n) #should be close to 1
print(np.mean(n))
print(np.std(n))
```
